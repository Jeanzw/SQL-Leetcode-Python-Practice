```python
import pandas as pd
```

```python
logs = pd.read_csv('Logs.csv')
logs
```

```python
logs['rank_num_by_id'] = logs.groupby('num')['id'].rank(method = 'first')
# 这一步就有点像是sql里面的row_number() over (partition by num order by id)的意思了
# logs.groupby('num') -> partition by num
# ['id'].rank(method = 'first') -> order by id，并且因为method = 'first'所以这个对应的是row number
logs['id - rank_num_by_id'] = logs['id'] - logs['rank_num_by_id']
# 形成新的一列，而这一列是就是我们之后判断是否是consecutive的来源了
logs
```

```python
logs_summary = logs.groupby(['num', 'id - rank_num_by_id']).count().reset_index().rename(columns={0: 'count_summary'})
# 这里如果没有后面的rename， 那么新生成的count的那一列其实是默认为0
logs_summary
```

```python
logs_summary = logs.groupby(['num', 'id - rank_num_by_id']).size().reset_index().rename(columns={0: 'count_summary'})
# 这里如果没有后面的rename， 那么新生成的count的那一列其实是默认为0
logs_summary
```
